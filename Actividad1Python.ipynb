3.- Programa en Python que lea 2 números y realice las 4 operaciones básicas

```python
num1=int(input("Ingresa el primer numero:"))
num2=int(input("Ingresa el segundo numero:"))
suma = num1+num2
resta = num1-num2
multiplicacion = num1*num2
division = num1/num2

print("La suma es:",suma)
print("La resta es:",resta)
print("La multiplicacion es:",multiplicacion)
print("La division es:",division)
```

5.- Programa en Python que lea una medida en centímetros y desplegar la misma medida pero convertida en pies y pulgadas

```python
centimetros=int(input("Ingresa una cantidad en centimetros:"))
pulgadas=(centimetros*0.393701)
pies=(centimetros*0.0328)
print("La medida en centimetros es:",centimetros)
print("La medida en pulgadas es:",pulgadas)
print("La medida en pies es:",pies)
```

6.- Programa en Python que lea una temperatura en grados centígrados y desplegarla en Grados Fahrenheit y grados kelvin

```python
temperaturaGradosCentigrados=int(input("Ingresa una temperatura en Grados Centigrados:"))
temperaturaGradosFarenheit=(temperaturaGradosCentigrados*1.8)+32
temperaturaGradosKelvin=temperaturaGradosCentigrados+273.15

print("Temperatura en Grados Centigrados:", temperaturaGradosCentigrados)
print("Temperatura en Grados Farenheit:", temperaturaGradosFarenheit)
print("Temperatura en Grados Kelvin:", temperaturaGradosKelvin)
```
